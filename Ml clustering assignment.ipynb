In this assignment students have to compress racoon grey scale image into
5 clusters. In the end, visualize both raw and compressed image and look
for quality difference.

```python
import numpy as np
from sklearn.cluster import KMeans
from scipy import misc
import matplotlib.pyplot as plt
```

```python
face=misc.face(gray=True)
```

```python
clusters=5
x=face.reshape((-1,1))
k_means=KMeans(n_clusters=clusters)
k_means.fit(x)
```

```python
labels=k_means.labels_
values=k_means.cluster_centers_
```

```python
compressed_face=np.choose(labels,values)
compressed_face.shape=face.shape
```

original image

```python
plt.imshow(face,cmap=plt.cm.gray)
```

compressed image

```python
plt.imshow(compressed_face,cmap=plt.cm.gray)
```
